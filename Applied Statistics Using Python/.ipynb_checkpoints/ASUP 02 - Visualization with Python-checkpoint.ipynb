### The Scatterplot

```python
x = [10, 15, 16, 23, 27, 38, 43, 56, 57, 60]
y = [5, 8, 9, 13, 16, 20, 40, 45, 67, 75]

plt.hist2d(x, y, bins=(50, 50), cmap=plt.cm.Reds);
```

```python
# Data for the 2D histogram
x = [10, 15, 16, 23, 27, 38, 43, 56, 57, 60]  # X-axis data
y = [5, 8, 9, 13, 16, 20, 40, 45, 67, 75]     # Y-axis data

# Create a 2D histogram
plt.figure(figsize=(6, 5))      # Set figure size
hist = plt.hist2d(
    x, y,                       # X and Y data points
    bins=(10, 10),              # Number of bins for X and Y axes
    cmap=plt.cm.Reds,           # Colormap for visualizing density
    cmin=1                      # Minimum count threshold for color display
)

plt.colorbar(hist[3], label="Frequency")  # Add a color bar

plt.xlabel("X-axis Values", fontsize=12)  # Label for the X-axis
plt.ylabel("Y-axis Values", fontsize=12)  # Label for the Y-axis

plt.title("2D Histogram of X and Y Data", fontsize=14, fontweight="bold") # Add a title to the plot

# Add gridlines for better readability
plt.grid(visible=True, linestyle="--", alpha=0.5)

plt.tight_layout()  # Adjust layout
plt.show()          # Display the plot
```

```python
# Data for the 2D histogram
x = [10, 15, 16, 23, 27, 38, 43, 56, 57, 60]  # X-axis data
y = [5, 8, 9, 13, 16, 20, 40, 45, 67, 75]     # Y-axis data

# Create a 1x2 grid of subplots
fig, axes = plt.subplots(1, 2, figsize=(12, 5))  # Set up a 1x2 subplot grid with larger figure size

# Plot the histogram with 50 bins on the first subplot
hist1 = axes[0].hist2d(x, y, bins=(50, 50), cmap=plt.cm.Reds)  # Plot the 2D histogram
fig.colorbar(hist1[3], ax=axes[0], label="Frequency")          # Add colorbar to the first subplot
axes[0].set_title("2D Histogram (bins=50)", fontsize=12)       # Add a title
axes[0].set_xlabel("X-axis Values", fontsize=10)               # Label for X-axis
axes[0].set_ylabel("Y-axis Values", fontsize=10)               # Label for Y-axis
axes[0].grid(True, linestyle='--', alpha=0.5)                  # Enable a grid for better readability

# Plot the histogram with 10 bins on the second subplot
hist2 = axes[1].hist2d(x, y, bins=(10, 10), cmap=plt.cm.Reds, cmin=1)  # Use fewer bins for visualization
fig.colorbar(hist2[3], ax=axes[1], label="Frequency")                  # Add colorbar to the second subplot
axes[1].set_title("2D Histogram (bins=10)", fontsize=12)               # Add a title
axes[1].set_xlabel("X-axis Values", fontsize=10)                       # Label for X-axis
axes[1].set_ylabel("Y-axis Values", fontsize=10)                       # Label for Y-axis
axes[1].grid(True, linestyle='--', alpha=0.5)                          # Enable a grid for better readability

plt.tight_layout()                                                     # Adjust layout
plt.show()                                                             # Display the final plots
```

### Histograms and Bar Graphs

```python
height = [10, 20, 30, 75, 100]
bars = ('A', 'B', 'C', 'D', 'E')

y_pos = np.arange(len(bars))
plt.bar(y_pos, height, color=(0.2, 0.4, 0.4, 0.6));
```

```python
# Data for the bar chart
height = [10, 20, 30, 75, 100]                 # Heights of the bars representing values
bars = ('A', 'B', 'C', 'D', 'E')               # Labels for each bar

y_pos = np.arange(len(bars))                   # Define the positions of the bars on the X-axis

colors = plt.cm.Set2(range(len(y_pos)))        # Define a color palette using Set2 colormap

# Create the bar chart
plt.figure(figsize=(6, 5))                     # Set figure size
plt.bar(
    y_pos,                                     # X-axis positions
    height,                                    # Heights of the bars
    color=colors,                              # Custom color palette Set2
    edgecolor='black',                         # Black edges for the bars for better definition
    alpha=0.9                                  # Set transparency level for the bars
)

# Add labels and title
plt.xticks(y_pos, bars, fontsize=10)                            # Set positions and labels for the X-axis
plt.xlabel('Categories', fontsize=12)                           # Label for the X-axis
plt.ylabel('Values', fontsize=12)                               # Label for the Y-axis
plt.title('Bar Chart Example', fontsize=14, fontweight='bold')  # Add title

plt.grid(axis='y', linestyle='--', alpha=0.7)  # Add gridlines

plt.tight_layout()                             # Adjust layout
plt.show()                                     # Display the bar chart
```

### Plotting Side-by-Side Histograms

```python
df = sns.load_dataset('iris')

sns.histplot( df["sepal_length"], color="skyblue", label="Sepal Length")
sns.histplot( df["sepal_width"], color="red", label="Sepal Width")

plt.show()
```

```python
# Load the Iris dataset
df = sns.load_dataset('iris')  # Load the built-in Iris dataset from seaborn

# Create a figure with a specific size
plt.figure(figsize=(6, 5))

# Create the histogram plot for sepal length
sns.histplot(
    data=df,                      # Use the dataframe as the data source
    x="sepal_length",             # Column to plot on the X-axis
    color="skyblue",              # Set the color of the bars for sepal length
    label="Sepal Length",         # Add a label for the legend
    alpha=0.7,                    # Set transparency for better overlap visibility
    kde=True                      # Add a Kernel Density Estimate (smooth line) for visualizing the distribution
)

# Create the histogram plot for sepal width
sns.histplot(
    data=df,                      # Use the dataframe as the data source
    x="sepal_width",              # Column to plot on the X-axis
    color="red",                  # Set the color of the bars for sepal width
    label="Sepal Width",          # Add a label for the legend
    alpha=0.7,                    # Set transparency for better overlap visibility
    kde=True                      # Add a Kernel Density Estimate (smooth line) for visualizing the distribution
)

# Add labels and a title to the plot
plt.xlabel("Measurement (cm)", fontsize=12)  # X-axis label
plt.ylabel("Frequency", fontsize=12)         # Y-axis label
plt.title("Distribution of Sepal Dimensions", fontsize=14, fontweight='bold')  # Title of the plot

plt.legend(fontsize=10, title="Features", title_fontsize=12)  # Add a legend
plt.grid(axis='y', linestyle='--', alpha=0.7)                 # Add gridlines

# Display the plot
plt.tight_layout()  # Adjust layout to prevent overlap
plt.show()          # Render the plot
```

### Pie Plots

```python
# Create a DataFrame with one column 'x' and custom index
df2 = pd.DataFrame(
    [8, 8, 1, 2],                # Data values
    index=['a', 'b', 'c', 'd'],  # Labels for the index
    columns=['Percentage']                # Column name
)

# Plot the data as a pie chart
df2.plot(
    kind='pie',                  # Specify pie chart
    subplots=True,               # Create a subplot for each column (useful for multiple columns)
    figsize=(5, 5),              # Set the figure size to 6x6 inches
    autopct='%1.1f%%',           # Display percentages on the chart with one decimal place
    legend=True ,                # Remove the legend to reduce clutter
    startangle=90,               # Start the pie chart at 90 degrees for consistency
    #labels=None,                 # Remove the index labels from the pie chart
    colors=plt.cm.Set2.colors    # Use a predefined color palette for the slices
)

# Add a title to the pie chart
plt.title("Actual Situation", fontsize=14, fontweight='bold')  # Title with formatting

# Adjust the layout
plt.tight_layout()

# Show the plot
plt.show()
```

### Heatmaps

```python
np.random.seed(18)

df3 = pd.DataFrame(np.random.random((5,5)), columns=['a','b','c','d','e'])

df3
```

```python
map = sns.heatmap(df3)
```

```python
# Generate a heatmap for the DataFrame 'df3'
plt.figure(figsize=(8, 6))   # Set the figure size for better readability
map = sns.heatmap(
    df3,                     # DataFrame to plot
    annot=True,              # Annotate each cell with its value
    fmt=".2f",               # Format for annotations (2 decimal places)
    cmap="coolwarm",         # Colormap for the heatmap
    cbar=True,               # Display the color bar
    linewidths=0.5,          # Add lines between cells for clarity
    linecolor="black"        # Color of the lines between cells
)

# Add labels and a title
plt.title("Heatmap of DataFrame 'df3'", fontsize=14, fontweight='bold')  # Title with formatting
plt.xlabel("Columns", fontsize=12)                                       # Label for the X-axis
plt.ylabel("Rows", fontsize=12)                                          # Label for the Y-axis

# Display the heatmap
plt.tight_layout() 
plt.show()
```

```python
df4 = sns.load_dataset('iris')

sns.jointplot(x=df4["sepal_length"], y=df4["sepal_width"], kind='hex')
sns.jointplot(x=df4["sepal_length"], y=df4["sepal_width"], kind='scatter')
sns.jointplot(x=df4["sepal_length"], y=df4["sepal_width"], kind='kde');
```

```python
# Load the Iris dataset
df4 = sns.load_dataset('iris')

# Create a figure and axes for a 1x3 grid
fig, axes = plt.subplots(1, 3, figsize=(18, 6))

# Plot 1: Hexbin plot
hb = axes[0].hexbin(
    x=df4["sepal_length"],  # X-axis data: sepal length
    y=df4["sepal_width"],   # Y-axis data: sepal width
    gridsize=30,            # Number of hexagons along each axis
    cmap="Blues"            # Use the "Blues" colormap
)
axes[0].set_title('Hexbin Plot', fontsize=14)  # Set a title for the first plot
axes[0].set_xlabel('Sepal Length')             # Label for the X-axis
axes[0].set_ylabel('Sepal Width')              # Label for the Y-axis
plt.colorbar(hb, ax=axes[0], label='Count')    # Add a color bar for the hexbin plot

# Plot 2: Scatter plot
sns.scatterplot(
    x="sepal_length",  # X-axis data: sepal length
    y="sepal_width",   # Y-axis data: sepal width
    data=df4,          # Use the Iris dataset as the data source
    ax=axes[1],        # Place the plot on the second subplot
    color='blue'       # Set the color of the points to blue
)
axes[1].set_title('Scatter Plot', fontsize=14)  # Set a title for the second plot
axes[1].set_xlabel('Sepal Length')              # Label for the X-axis
axes[1].set_ylabel('Sepal Width')               # Label for the Y-axis

# Plot 3: KDE plot
sns.kdeplot(
    x=df4["sepal_length"],  # X-axis data: sepal length
    y=df4["sepal_width"],   # Y-axis data: sepal width
    ax=axes[2],             # Place the plot on the third subplot
    cmap="Blues",           # Use the "Blues" colormap for shading
    fill=True               # Fill the density areas with color
)
axes[2].set_title('KDE Plot', fontsize=14)  # Set a title for the third plot
axes[2].set_xlabel('Sepal Length')          # Label for the X-axis
axes[2].set_ylabel('Sepal Width')           # Label for the Y-axis

# Adjust layout and spacing
plt.tight_layout()  # Automatically adjust subplot parameters to prevent overlapping elements
plt.show()          # Display all the plots
```

### Line Charts

```python
np.random.seed(18)

values=np.cumsum(np.random.randn(1000,1))

plt.plot(values);
```

```python
# Set the random seed for reproducibility
np.random.seed(18)

# Generate a cumulative sum of 1000 random values
values = np.cumsum(np.random.randn(1000, 1))

# Create the plot
plt.figure(figsize=(8, 5))  
plt.plot(values, color='blue', linewidth=1.5, label='Cumulative Sum') 

# Add title and labels
plt.title('Cumulative Sum of Random Values', fontsize=14, fontweight='bold')  # Add a title to the plot
plt.xlabel('Steps', fontsize=12)                                              # Label the X-axis as "Steps"
plt.ylabel('Value', fontsize=12)                                              # Label the Y-axis as "Value"

plt.grid(visible=True, linestyle='--', alpha=0.7)   # Add gridlines

plt.legend(fontsize=10, loc='upper left')  # Add a legend

plt.tight_layout()                         # Adjust the layout
plt.show()                                 # Display the plot
```
